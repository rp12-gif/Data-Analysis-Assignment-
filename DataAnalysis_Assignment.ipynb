```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Marketing_Design_Dataset.csv")
```

```python
df.head(4)
```

```python
print(df.describe())
```

```python
plt.hist(df["Campaign_Budget"], bins=20, color="pink", alpha=0.7)
plt.xlabel("Campaign Budget ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Campaign Budget")
plt.show()
```

```python
plt.scatter(df["Ad_Click_Rate"], df["Conversion_Rate"], color="pink")
plt.xlabel("Ad Click Rate (%)")
plt.ylabel("Conversion Rate (%)")
plt.title("Ad Click Rate vs. Conversion Rate")
plt.show()
```

```python
df["Customer_Retention_Rate"].plot(kind="bar", color="pink")
plt.xlabel("Index")
plt.ylabel("Retention Rate (%)")
plt.title("Customer Retention Rate Distribution")
plt.show()
```
